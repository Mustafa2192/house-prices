```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn import preprocessing
import seaborn as sns
from sklearn.linear_model import LinearRegression

warnings.filterwarnings('ignore')
plt.style.use("seaborn-whitegrid")
%matplotlib inline
```

```python
trainData = pd.read_csv('data/train.csv')
```

```python
print("Training Data Size: {}".format(len(trainData)))
```

```python
trainData.shape
```

## Impute Values

```python
len(trainData[trainData["MasVnrArea"].isnull()])
```

```python
len(trainData[trainData["GarageYrBlt"].isnull()])
```

```python
len(trainData[trainData["GarageYrBlt"].isnull() & trainData["GarageType"].notnull()])
```

```python
trainData["GarageYrBlt"].fillna(trainData["GarageYrBlt"].median(), inplace = True)
```

```python
trainData["MasVnrArea"].fillna(trainData["MasVnrArea"].median(), inplace = True)
```

## Data Plots

```python
X = trainData[trainData['GarageYrBlt']!=0][['GarageYrBlt']].values
X2 = trainData[trainData['YearBuilt']!=0][['YearBuilt']].values
y = trainData[trainData['GarageYrBlt']!=0][['SalePrice']].values

plt.figure(figsize=(16,6))
ax = plt.subplot()
ax.scatter(X,y,s=20, alpha=0.2, c='blue')
ax.scatter(X2,y,s=20, alpha=0.2, c='red')
ax.set_xlabel('GarageYrBlt')
ax.set_ylabel('SalePrice')
ax.legend(('GarageYrBlt','YearBuilt'),scatterpoints=1,loc='upper left',fontsize=12)
```

```python
X = trainData[trainData['GrLivArea']!=0][['GrLivArea']].values
y = trainData[trainData['GrLivArea']!=0][['SalePrice']].values
plt.figure(figsize=(16,6))
ax = plt.subplot()
ax.scatter(X,y,s=20, alpha=0.2)
ax.set_xlabel('GarageYrBlt')
ax.set_ylabel('SalePrice')
ax.legend(('GarageYrBlt','YearBuilt'),scatterpoints=1,loc='upper left',fontsize=12)

slr = LinearRegression()
slr.fit(X,y)
plt.plot(X,slr.predict(X),color='darkgray', alpha=0.8)
print('coeff：{0}'.format(slr.coef_[0]))
print('slope: {0}'.format(slr.intercept_))
```
